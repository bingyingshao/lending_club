#### Find and filter out the features that have more 50% of missing value

```python
missing = pd.DataFrame(loan_15.isnull().sum()/len(loan_15),columns=['%missing_value']).sort_values(by='%missing_value',ascending=False)
```

```python
missing
```

```python
feat = missing[missing['%missing_value']<0.5].index.tolist()
```

```python
len(feat)
```

```python
df = loan_15[feat].reset_index(drop=True)
del loan_15
```

```python
df = df[df.loan_status!='Current'].reset_index(drop=True)
```

```python
plt.figure(figsize=(25,5)) 
sns.set_context(font_scale=10)   
sns.heatmap(df.isnull(),yticklabels=False,cbar=False,cmap='viridis',xticklabels=list(range(88)))
```
